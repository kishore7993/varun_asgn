```python
# Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import skew, kurtosis

# Load dataset
df = pd.read_csv("supermarket_sales.csv")

# Display dataset information
df.info()
df.head()
```

```python
# Select numeric columns for analysis
numeric_df = df.select_dtypes(include=['number'])
```

```python
# Display dataset info
print("Dataset Info:")
df.info()
print("\nSummary Statistics:")
print(df.describe())
```

```python
# 1. Relational Plot: Scatter Plot - Unit Price vs Quantity
plt.figure(figsize=(10, 6))
sns.scatterplot(x=df['Unit price'], y=df['Quantity'], hue=df['Product line'], alpha=0.7)
plt.title('Scatter Plot: Unit Price vs Quantity')
plt.xlabel('Unit Price')
plt.ylabel('Quantity')
plt.legend(loc='upper right')
plt.show()
```

```python
# 2. Categorical Plot: Bar Plot - Payment Methods
plt.figure(figsize=(8,5))
sns.countplot(x=df['Payment'], palette='viridis')
plt.title('Bar Plot: Payment Methods')
plt.xlabel('Payment Method')
plt.ylabel('Count')
plt.show()
```

```python
# 3. Statistical Plot: Correlation Heatmap
plt.figure(figsize=(10,6))
sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Heatmap')
plt.show()
```

```python
# 4. Statistical Plot: Box Plot - Total Sales by Product Line
plt.figure(figsize=(12,6))
sns.boxplot(x=df['Product line'], y=df['Total'], palette='coolwarm')
plt.title('Box Plot: Total Sales by Product Line')
plt.xlabel('Product Line')
plt.ylabel('Total Sales')
plt.xticks(rotation=45)
plt.show()
```

```python
# 5. Statistical Moments - Mean, Variance, Skewness, Kurtosis
feature = 'Total'

mean_val = df[feature].mean()
variance_val = df[feature].var()
skewness_val = skew(df[feature])
kurtosis_val = kurtosis(df[feature])

print(f"\nStatistical Moments for {feature}:")
print(f"Mean: {mean_val}")
print(f"Variance: {variance_val}")
print(f"Skewness: {skewness_val}")
print(f"Kurtosis: {kurtosis_val}")
```
